```python
import matplotlib.pyplot as plt
import numpy as np
import time

import utils
import histogram

hist_figsize = (25.6, 5)
testcases = 100
```

# Histogram Equalization
### Q3_1_1

```python
img_raw = utils.loadGrayscaleAsNumpy("res/Q3_1_1.tif")
start = time.time()
for i in range(testcases): img, img_hist, raw_hist = histogram.hist_equ(img_raw)
print("Average Time Cost: {} s.".format((time.time() - start) / testcases))

plt.figure(1, figsize=hist_figsize)
utils.plotHistogram(plt, raw_hist, "lightgray")
utils.plotHistogram(plt, img_hist, "orange")

plt.figure(2)
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
utils.saveGrayscale(img, "output/output_Q3_1_1_12112907.tif")

plt.show()
```

# Histogram Matching

```python
img_raw = utils.loadGrayscaleAsNumpy("res/Q3_2.tif")

spec_hist = np.concatenate((
    np.linspace(0, 7, 5 - 0 + 1),
    np.linspace(7, 0.7, 20 - 5 + 1)[1:],
    np.linspace(0.7, 0, 181 - 20 + 1)[1:],
    np.linspace(0, 0.6, 203 - 181 + 1)[1:],
    np.linspace(0.6, 0, 255 - 203 + 1)[1:]
), axis=0)
spec_hist /= np.sum(spec_hist)

start = time.time()
for i in range(testcases): img, img_hist, raw_hist = histogram.hist_match(img_raw, spec_hist)
print("Average Time Cost: {} s.".format((time.time() - start) / testcases))

plt.figure(1, figsize=hist_figsize)
utils.plotHistogram(plt, spec_hist, "blue")

plt.figure(2, figsize=hist_figsize)
utils.plotHistogram(plt, raw_hist, "lightgray")
utils.plotHistogram(plt, img_hist, "orange")

plt.figure(3)
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
utils.saveGrayscale(img, "output/output_Q3_2_12112907.tif")

plt.show()
```
