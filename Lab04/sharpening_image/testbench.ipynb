# Sharpening Image

```python
import matplotlib.pyplot as plt
import numpy as np

import utils
import sharpening
import time
```

### Q4_1

```python
img_raw = utils.loadGrayscaleAsNumpy("res/Q4_1.tif")
start = time.time()
img = sharpening.combined_spatial_sharpening(img_raw, 0.8)
print("Average Time Cost: {} s.".format((time.time() - start)))

plt.figure()
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
utils.saveGrayscale(img, "output/output_Q4_1_12112907.tif")

plt.show()
```
