```python
import matplotlib.pyplot as plt
import time

import utils
import interpolation

img_raw = utils.loadGrayscaleAsNumpy("res/rice.tif")
scalar = 4
```

# Nearest
### v0

```python
start = time.time()
img = interpolation.nearest_v0(img_raw, (round(256 * scalar), round(256 * scalar)))
print("Time Cost: {} s.".format(time.time() - start))

plt.imshow(img, cmap="gray")
plt.show()
```

### v1

```python
start = time.time()
img = interpolation.nearest_v1(img_raw, (round(256 * scalar), round(256 * scalar)))
print("Time Cost: {} s.".format(time.time() - start))

plt.imshow(img, cmap="gray")
plt.show()
```

### latest

```python
start = time.time()
img = interpolation.nearest(img_raw, (round(256 * scalar), round(256 * scalar)))
print("Time Cost: {} s.".format(time.time() - start))

plt.imshow(img, cmap="gray")
plt.show()

utils.saveGrayscale(img, "output/enlarged_nearest_12112907.tif")
```
